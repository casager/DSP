# Introduction Code Sample 1

This code is a simple simulation of Example 2.11 from our Textbook "Discrete Time Signals & Systems" by Oppenheim and Schafer. In this code we will be simulating a system with impulse response 
$$
h[n] = u[n] - u[n-N]
$$

In theory this code will work with any input, x, but we will be simulating 
$$
x[n] = a^n*u[n]
$$
In addition, and something I forgot while going over this example during class, we will assume 0 < a < 1

Since we will be applying this system impulse response to the input to get the output, we will be performing a convolution sum.
There are built in functions to perform this; however, we will write out own simple one for today.

Remember the equation for a convolution sum is

$$
y[n] = \sum\limits _{k=-\infty} ^{\infty} x[k]h[n-k]
$$


## Code Setup
To run this code, we will need a few built in python libraries to make our lives easier

```python
import numpy as np
import matplotlib.pyplot as plt #this is how you will probably prefer to plot
```

## Code Planning and variable creating

Now that we have our libraries imported, we will refer to them as their shortened names in the code. Next, we need to 
<ol>
    <li>Create variables for defining our impulse response, our input, and our output. Most importantly we will need the length of these arrays, and the value of N </li>
    <li> create our arrays for our impulse response and our input </li>
</ol> 

```python
N = 10 #length of the impulse response
a = 0.95 #value of a in our input
simulationLength = 100 #how large of an array we want to simulate for the input and the output

n = np.arange(simulationLength) # this creates an array of length simulation length increasing from 0 to simulationLength - 1
h = np.ones(N) #creates an array of ones of length N. This is our impulse response
y = np.zeros(simulationLength) #creates an array of zeros to store the output

x = np.power(a, n) #this creates our input. Note: just like matlab, numpy lets us avoid the for loop to create this array

plt.figure("Input")
plt.stem(x)
plt.show()
```

## The actual algorithm

This isn't necessarily written in the most computationally simple way (hint: it isn't) and it isn't a generic solution for convolution. Instead this is implementing the system defined above in a simple to understand method.

Pay attention to some of the decisions I made in this algorithm as it avoids 

```python
for i in range(len(y)):
    samplesToSum = N
    if i < N:
        samplesToSum = i
    for j in range(samplesToSum):
        y[i] = y[i] + x[i - j]*h[N-1-j]
plt.figure('Output')
plt.stem(y)
plt.show()
```
